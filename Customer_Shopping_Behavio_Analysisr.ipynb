```python
import pandas as pd
df = pd.read_csv('/content/customer_shopping_behavior.csv')
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create column age_grup
labels = ['Young Adult', 'Adult','Middle_Age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequently_days
frequency_mapping = {
    'Forthnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quaterly': 90,
    'By_Weekly':14,
    'Annually': 365,
    'Every 3 Months':90
}

df['purchase_frequently_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequently_days','frequency_of_purchases']].head(10)
```
